```python
import math
```

```python
def function1(x):
    temp1 = pow(x, 4)
    temp2 = math.exp(-2*x)
    return temp1*temp2
```

```python
def function2(x):
    temp1= pow(x, 2)*math.exp(-1*x)*math.sin(x**2)
    return temp1**2
```

```python
x = np.arange(0,8, 0.01 )
y = np.array(list(map(function1, x)))
z  = np.array(list(map(function2, x)))

fig = plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.plot(x,y)
plt.xlim(0,7)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()

plt.subplot(122)
plt.plot(x,z)
plt.xlim(0,7)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
#position function

def position(t):
    return -0.5*9.8*t**2
def velocity(t):
    return -9.8*t
def acceleration(t):
    return 9.8
t = np.arange(0, 10, 0.1)
x = np.array(list(map(position, t)))
v= np.array(list(map( velocity, t)))
a = np.array(list(map(acceleration, t)))
```

```python


fig, ax = plt.subplots()

ax.set_title("Graphs of positions., velocities, accelerations")
ax.plot(t, x)

ax2 = ax.twinx()

ax2.plot(t, v)

ax3 = ax2.twinx()

ax3.plot(t, a)

plt.show()

```
